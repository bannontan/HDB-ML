## Adjusting for inflation using CPI ##

```python
cpi = pd.read_csv("./data/CPI.csv")
cpi.head()
```

```python
# converts cpi["month"] data from %Y %b format to datetime format
cpi["month"] = pd.to_datetime(cpi["month"], format = "%Y %b")
cpi.head()
```

```python
# merge prices dataframe with cpi dataframe
# it merges on = "month". This means that the dataframe merge together based on the values of month. If the month value is the same then they will merge.
# how = "left" means that all the information from the left dataframe, in this case prices, will be preserved in the final dataframe

prices = prices.merge(cpi, on = "month", how = "left")
```

```python
# finding the real price of the hdb at the reference period
# formula is: cost at reference period = (cost in current period / cpi) * 100

prices["real_price"] = (prices["resale_price"] / prices["cpi"]) * 100
prices.head()
```

## Data cleaning / Preprocessing ##

```python
# function that converts text to float for years for remaining_lease. eg: if it is 1 year and 6 months then return 1.5
# this helps to ensure that the expression of remaining_lease is consistent and enables us to calculate statistics like mean and median

def getYears(text):
    if isinstance(text, str):
        yearmonth= [int(s) for s in text.split() if s.isdigit()]

        if len(yearmonth) > 1:
            years = yearmonth[0] + yearmonth[1] / 12
        else:
            years = yearmonth[0]

        return years
    else: # if text is an int/float
        return text
    
prices["remaining_lease"] = prices["remaining_lease"].apply(lambda x: getYears(x))
prices.info()
```

## Machine Learning Model ##

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
# split the data into feature and target
x = prices.drop("resale_price", axis=1)
y = prices["resale_price"]

# split the data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# scale the data
# MinMaxScaler scales features to a specified range, typically between 0 and 1, by subtracting the minimum value and then dividing by the difference between the maximum and minimum values
# scaler = MinMaxScaler()
# x_train = scaler.fit_transform(x_train)  # calculates the minimum and maximum values for each feature in the training data and then transforms (scales) the training data using these values
# x_test = scaler.transform(x_test)  # applies the same transformation that was learned from the training data to the testing data
```

```python
# create the model
model = LinearRegression()
model.fit(x_train, y_train)

print("Linear Regression Train Score: ", model.score(x_train, y_train))
print("Linear Regression Test Score: ", model.score(x_test, y_test))
```

```python
# Lasso Regression
lasso_model = Lasso(alpha=0.5, max_iter=10000)
lasso_model.fit(x_train, y_train)

# Model Evaluation
print("Lasso Train Score: ", lasso_model.score(x_train, y_train))
print("Lasso Test Score: ", lasso_model.score(x_test, y_test))
```

```python
# Ridge Regression
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(x_train, y_train)

# Model Evaluation
print("Ridge Train Score: ", ridge_model.score(x_train, y_train))
print("Ridge Test Score: ", ridge_model.score(x_test, y_test))
```
